```python
#import necessary libraries
import os
from tqdm import tqdm
```

```python
#integer to store the month count
month = 12

#input hashes list
input_hashes = []

#read all input files
for i in tqdm(range(month)):
    filename = "./Data/edges2010/inputs2010_{}.txt".format(str(i+1))
    
    with open(filename) as file:
        for line in file:
            #split by tab char to get the details present
            details = line.split("\t")
            
            #hash present at the second value
            #append in the list
            input_hashes.append(details[1])
```

```python
print("Total unique input hashes are: {}".format(len(input_hashes)))
```

```python
#output hashes list
output_hashes = []

#read all output files
for i in tqdm(range(month)):
    filename = "./Data/edges2010/outputs2010_{}.txt".format(str(i+1))
    
    with open(filename) as file:
        for line in file:
            #split by tab char to get the details present
            details = line.split("\t")
            
            #hash present at the second value
            #append in the list
            output_hashes.append(details[1])
```

```python
print("Total unique output hashes are: {}".format(len(output_hashes)))
```

```python
#create mapping
mapping = dict(zip(set(output_hashes), range(len(output_hashes))))
```

```python
#create a updated folder to store the updated text files
folder_name = "./Data/edges2010/updated"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)
```

```python
#get the updated output files
for i in tqdm(range(month)):
    filename = "./Data/edges2010/outputs2010_{}.txt".format(str(i+1))
    updated_filename = "./Data/edges2010/updated/outputs2010_{}.txt".format(str(i+1))
    
    file = open(filename, 'r')
    contents = file.read()
    
    for key, value in mapping.items():
        contents = contents.replace(key, str(value))
    
    updated_file = open(updated_filename, 'w')
    updated_file.write(contents)
    
    file.close()
    updated_file.close()
```
